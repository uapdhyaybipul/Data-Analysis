```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['Grade'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.rename(columns={'Min_x0020_Price':'Min_Price','Max_x0020_Price':'Max_Price'},inplace=True)
```

```python
df
```

```python
df['Max_Price'].fillna(df['Max_Price'].mean(),inplace=True)
df['Max_Price'].isna().sum()
```

```python
df['Min_Price'].fillna(df['Min_Price'].mean(),inplace=True)
df['Min_Price'].isna().sum()
```

```python
sns.scatterplot(data=df,x='Min_Price',y='Modal_x0020_Price')
plt.show()
```

```python
df['Revenue']=df['Min_Price']+df['Max_Price']+df['Modal_x0020_Price']
comm=df.groupby('Commodity')[['Min_Price','Max_Price','Modal_x0020_Price','Revenue']].sum().sort_values(by='Revenue',ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(x=comm.index,height=comm['Revenue'],color=['purple','cyan','green','red','pink','brown','olive','cyan'])
plt.xticks(rotation=75)
plt.title('India Top 10 revenue generated commodity')
plt.show()
```
