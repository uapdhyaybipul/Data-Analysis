## Graphs and Charts

**1.] Histogram**

The histogram displays a frequency distribution (shape) of a data set. At first glance, histograms look alike to bar graphs. However, there is a key difference between them. Bar Chart represents categorical data and histogram represent continuous data.
ers.

1.When the data is continuous.


2.When you want to represent the shape of the datas distribution.


3.When you want to see whether the outputs of two or more processes are different.


4.To summarize large data sets graphically.


5.To communicate the data distribution quickly to others.

```python
x=np.linspace(2,80,15)
plt.xticks(x,rotation='vertical')
plt.hist(x)
plt.show()
```

**2.] Bar Charts**


Bar charts represent categorical data with rectangular bars. Bar graphs are among the most popular types of graphs and charts in economics, statistics, marketing, and visualization in digital customer experience. They are commonly used to compare several categories of data.

```python
labels=['car','Bus','Truck','Bike']
x_pos=np.arange(len(labels))
users=[20,120,45,70]

plt.xticks(x_pos,labels)
plt.title('Vehicle Users',fontname='Comic Sans MS',fontsize=18)
plt.xlabel('Vehicles',fontname='Comic Sans MS',fontsize=15)
plt.ylabel('No. of Users',fontname='Comic Sans MS',fontsize=15)
plt.bar(x_pos,users,color='green')
plt.show()
```

## Pie Charts

A pie chart (or a circle chart) is a circular statistical graphic which is divided into slices to illustrate numerical proportion.

legend position


```python
labels=['car','Bus','Truck','Bike']

users=[20,120,45,70]

explode = (0.1, 0.2, 0.1, 0.1)


plt.title('Vehicle Users',fontname='Comic Sans MS',fontsize=18)

plt.pie(users,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%')
plt.legend(loc=4,bbox_to_anchor=(0.8, 0.0, 0.5, 0.5))            ###bbox_to_anchor=(0.8, 0.0, 0.5, 0.5) for shiftin of legend

plt.show()
```
