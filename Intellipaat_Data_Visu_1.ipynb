```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cars=pd.read_csv('mtcars.csv')
df_cars.head()
```

```python
df_cars.rename(columns={'Unnamed: 1':'model'},inplace=True)
```

```python
sns.lineplot(data=df_cars,x=df_cars['model'],y=df_cars['hp'])
plt.xlabel('Model')
plt.ylabel('hp')
plt.xticks(rotation=90)
plt.title('Model vs hp')
plt.show()
```
